```python
#Conversion of Inches into float datatype

inc=df.Inches
inc1=[]
for i in range (1303):
  inc1.append(float(inc[i]))
```

```python
#Conversion of ram into float type

ram2=df.Ram
row2=[]
for i in range(1303):
  row1=ram2[i].split("GB").pop(0)
  row1=int(row1)
  row2.append(row1)
```

```python
#conversion of memory into float type

memory2=df.Memory
row4=[]
for i in range(1303):
  row1=memory2[i].split(" ")
  try:
    if 'SSD' in row1:
      row1.remove('SSD')
    if 'Flash' in row1:
      row1.remove('Flash')
    if 'Storage' in row1:
      row1.remove('Storage')
    if 'HDD' in row1:
      row1.remove('HDD')
    row1=''.join(map(str,row1))
    if len(row1)==3:
      row1='1024'
    if len(row1)==4:
      row1=row1[:2]
    if len(row1)==5:
      row1=row1[:3]
    if len(row1)==9:
      row1='1152'
    if len(row1)==14:
      row1='512'
    if len(row1)==11:
      row1='1024'
    if len(row1)==8:
      row1='1088'
    if len(row1)==10:
      row1='2048'
    if len(row1)==7:
      row1='2048'
    if len(row1)==17:
      row1='1536'

  finally:
    row1=float(row1)
    row4.append(row1)
```

```python
#Conversion of weight into float type

Weight2=df.Weight
row6=[]
for i in range(1303):
  row1=Weight2[i].split("kg").pop(0)
  row1=float(row1)
  row6.append(row1)
```

```python
#Making the input data
x=pd.DataFrame(list(zip(inc1,row2,row4,row6)),columns=['Inches','Ram','Memory','Weight'])
x=pd.DataFrame(x)
print(x)
```

```python
import numpy
#Making the required output
y=df.iloc[:,12]
print(y)
```

```python
#MAKING Train and Test variables

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0) #random_state = 0 is to avoid duplicates

#By default Training variables get 75% of data and Testing variables get the remaining 25% of the data.
```

```python
print(x.shape) #100%
print(x_train.shape) #75%
print(x_test.shape) #25%
#print(y.shape) #100%
#print(y_train.shape) #75%
#print(y_test.shape) #25%
```

```python
#Normalisation of all variables
```

```python
#Applying a CLASSIFIER
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model (Mapping the values of x_train and y_train in the LogisticRegression library)
model.fit(x_train,y_train)
```

```python
# Predicting the output
y_pred = model.predict(x_test) #using the input testing values , we predict the output
y_pred #Predicted output values
```

```python
y_test #Actual output values
```

```python
#CONCLUSION : Wehave to compare both y_pred and y.
#SO when we compare the corresponding values of Y-pred and y , we come to know that there is huge difference.
#This huge difference does not mean that our model has predicted wrong.
#It only means that , our model is NOT LINEAR/LESS LINEAR.
#LINEARITY of a MODEL depends on the NATURE of the data as well as the size of the data.
```

```python
#Finding the accuracy of our model

from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print("The accuracy of our model is {}%".format(round(score, 2) *100))
```
